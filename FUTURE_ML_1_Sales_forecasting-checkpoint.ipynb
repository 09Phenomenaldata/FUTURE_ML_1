```python
# AI-Powered Sales Forecasting Dashboard - Improved Version

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from prophet import Prophet
from sklearn.metrics import mean_absolute_error, mean_squared_error
import seaborn as sns

# Step 1: Generate synthetic sales data
def generate_sales_data(start='2024-01-01', periods=180):
    np.random.seed(42)
    date_range = pd.date_range(start=start, periods=periods, freq='D')
    trend = np.linspace(500, 600, periods)
    seasonality = 50 * np.sin(2 * np.pi * date_range.dayofyear / 30)
    noise = np.random.normal(0, 20, periods)
    sales = trend + seasonality + noise
    return pd.DataFrame({'ds': date_range, 'y': sales})

df = generate_sales_data()

# Step 2: Visualize sales
def plot_sales(df):
    plt.figure(figsize=(12, 5))
    sns.lineplot(x='ds', y='y', data=df, label='Sales', color='blue')
    plt.title(' Retail Sales Over Time')
    plt.xlabel('Date')
    plt.ylabel('Sales')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

plot_sales(df)

# Step 3: Train/Test Split
train = df[:-30]
test = df[-30:]

# Step 4: Train Prophet Model
model = Prophet(
    yearly_seasonality=False,
    weekly_seasonality=True,
    daily_seasonality=False,
    seasonality_mode='additive',
    changepoint_prior_scale=0.1
)
model.fit(train)

# Step 5: Make future predictions
future = model.make_future_dataframe(periods=30)
forecast = model.predict(future)

# Step 6: Plot forecast
def plot_forecast(model, forecast):
    fig = model.plot(forecast)
    plt.title(" 30-Day Sales Forecast")
    plt.xlabel("Date")
    plt.ylabel("Sales")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

plot_forecast(model, forecast)

# Step 7: Plot components
def plot_components(model, forecast):
    model.plot_components(forecast)
    plt.tight_layout()
    plt.show()

plot_components(model, forecast)

# Step 8: Evaluate Model
forecast_eval = forecast.set_index('ds')[['yhat']].join(test.set_index('ds'))
forecast_eval.dropna(inplace=True)

def evaluate_forecast(df_eval):
    mae = mean_absolute_error(df_eval['y'], df_eval['yhat'])
    rmse = np.sqrt(mean_squared_error(df_eval['y'], df_eval['yhat']))
    print(f" Mean Absolute Error (MAE): {mae:.2f}")
    print(f" Root Mean Squared Error (RMSE): {rmse:.2f}")
    return mae, rmse

mae, rmse = evaluate_forecast(forecast_eval)

# Step 9: Actual vs Predicted Plot
def plot_actual_vs_predicted(df_eval):
    plt.figure(figsize=(12, 5))
    plt.plot(df_eval.index, df_eval['y'], label='Actual Sales', marker='o')
    plt.plot(df_eval.index, df_eval['yhat'], label='Predicted Sales', linestyle='--')
    plt.title(" Actual vs Predicted Sales (Last 30 Days)")
    plt.xlabel("Date")
    plt.ylabel("Sales")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Merge actual sales into comparison dataframe
forecast_eval['y'] = test.set_index('ds')['y']
plot_actual_vs_predicted(forecast_eval)
```
